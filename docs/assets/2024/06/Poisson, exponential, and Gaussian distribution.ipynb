## Three distributions

Notes on 01.11.2024

The Poisson distribution is a discrete distribution. It expresses the probability of a give number of events occuring in a fixed interval of time, if these events occur with a known, constant mean rate, and independently of the time since the last event. The shape of a Poisson distribution is determined by the expectation of number of events in a given time interval, which is usually noted as $\lambda$. Given the expectation $\lambda$, the probability of observing $k$ ($k\ge0$) events in the intervalis

$$ \frac{\lambda^{k}e^{-\lambda}}{k!} $$

When $\lambda$ is large, or when the time interval is large, the Poisson distribution resembles the Gaussian distribution.

A Poisson point process is a process in which events occur continuously and independently at a constant avarage rate $\lambda$. 

The exponential distribution is the probability distribution of the distance between events in a Poisson point process. It can be used to model the time between events that occur continuously and independently at a constant average rate, for instance production errors, or natural births. Given the rate $\lambda$, the probability distribution is written as

$$ \lambda e^{-\lambda x} \quad(x\ge0)$$


```python
from matplotlib import pyplot as plt
import numpy as np

np.random.seed(1887)

time = range(1, 50)
events_int = [3, 9, 17, 19, 27, 34, 39, 44]
events =np.random.normal(loc=events_int, scale=0.8)

plt.hlines(1,1,50, color="black")  # Draw a horizontal line
plt.xlim(0,51)
plt.ylim(0.8,1.2)

y = np.ones(np.shape(time))   # Make all y values the same
events_y = [1.02] * len(events)

plt.plot(time,y,'|',ms = 5)  # Plot a line at each location specified in a
plt.plot(events, events_y, 'v')
plt.axis('off')

poisson_color = "#004495"
exponential_color = "#AA3800"
gaussian_color = "#AA00BB"

plt.arrow(1, 1, 49, 0, head_width=0.01, head_length=1, ec="black", fc="black")
plt.text(50, 1.01, "time", fontstyle="italic")

plt.arrow(1, 1.07, 1, 0, head_width=0.01, head_length=0.01, ec=poisson_color)
plt.arrow(2, 1.07, -1, 0, head_width=0.01, head_length=0.01, ec=poisson_color)
plt.text(1, 1.08, "Histogram of #events in each time unit: Poisson distribution", color=poisson_color)

plt.arrow(events[2], 1.03, events[3]-events[2], 0, head_width=0.01, head_length=0.01, ec=exponential_color)
plt.arrow(events[3], 1.03, events[2]-events[3], 0, head_width=0.01, head_length=0.01, ec=exponential_color)
plt.text(1, 1.04, "Histogram of time intervals between events: Exponential distribution", color=exponential_color)

plt.arrow(1, 0.97, 49, 0, head_width=0.01, head_length=0.01, ec=gaussian_color)
plt.arrow(49, 0.97, -48, 0, head_width=0.01, head_length=0.01, ec=gaussian_color)
plt.text(1, 0.95, "Histogram of #events in large time internvals: Gaussian distribution", color=gaussian_color)

plt.title("Relationship between Gaussian, Poisson, and exponential distributions")
plt.show()
```
